- Po przebadaniu I i III ćwiartki można dojśc do wniosku, że dane z ramki czwartej sa nijak skorelowane z pozostałymi danymi, a ewentualne zależności wynikają jedynie z połączenia z bardzo dobrą cechą (patrz. meanfun)
- Fajna jest jeszcze IQR.
- Można się przyjrzeć Q25, sd.

```python
d4 = d4.drop('label', axis=1)

d = d.drop(d4.columns, axis=1)
d.head()
```

```python
df = pd.concat([d.loc[:,'IQR'], d.loc[:,'Q25'], d.loc[:,'sd'], d.loc[:,'meanfun'], d.loc[:, 'meanfreq'], d.loc[:,'sp.ent'],d.loc[:,'sfm'], d.loc[:,'label']], axis=1)
```

```python
df.head()
```

```python
sns.pairplot(df, kind="scatter", hue = "label")
plt.figure(figsize=(10,5), dpi= 80)

plt.show()
```

```python
#IQR, Q25
atr = 'meanfun'
plt.figure(figsize=(10,7), dpi= 80)
sns.distplot(df.loc[df.label == 'male',atr], label="Male")
sns.distplot(df.loc[df.label == 'female',atr], label="Female")

plt.legend();
```
